```python
import pandas as pd
import numpy as np
from scipy.stats import linregress
from scipy import stats
import matplotlib.pyplot as plt
```

```python
cities_df = pd.read_csv("../Resources/cities.csv")
cities_df.head(10)
```

```python
#Temperature section
plt.scatter(cities_df['Lat'], cities_df['Max Temp'], c='blue')
plt.xlabel("Latitude")
plt.ylabel("Max Temperature(F)")
plt.savefig("../Resources/assets/images/MyFig1.png")
```

```python
#Humditity section
plt.scatter(cities_df['Lat'], cities_df['Humidity'], c='blue')
plt.xlabel("Latitude")
plt.ylabel("Humidity(%)")
plt.savefig("../Resources/assets/images/MyFig2.png")
```

```python
#Cloudiness section
plt.scatter(cities_df['Lat'], cities_df['Cloudiness'], c='blue')
plt.xlabel("Latitude")
plt.ylabel("Cloudiness(%)")
plt.savefig("../Resources/assets/images/MyFig3.png")
```

```python
#Wind Speed section
plt.scatter(cities_df['Lat'], cities_df['Wind Speed'], c='blue')
plt.xlabel("Latitude")
plt.ylabel("Wind Speed(mph)")
plt.savefig("../Resources/assets/images/MyFig4.png")
```
